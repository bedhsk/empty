```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Cargamos los datos del titanic

```python
data = sns.load_dataset('titanic')
data.head(10)
```

## Sustitución de valores nulos en la columna edad

```python
# Obtendremos el promedio y su desviación estandar
edad_media = data['age'].mean()
des_est =data['age'].std()
cuenta_edad_nula = data['age'].isnull().sum()
lista_edad_nula = np.random.randint(edad_media - des_est, edad_media + des_est, size=cuenta_edad_nula)

con_valores_nulos = np.isnan(data['age'])
data.loc[np.isnan(data['age']), 'age'] = lista_edad_nula
data['age'] = data['age'].astypepe(int)

print("Edad Promedio: " + str(edad_media))
print("Desvió Std Edad: " + str(des_est))
print("Intervalo para asignar edad aleatoria: " + str(int(edad_media - des_est)) + " a " + str(int(edad_media + des_est)))
```
